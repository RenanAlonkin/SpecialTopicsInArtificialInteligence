# Conjunto de dados IRIS

```python
df = sns.load_dataset('iris')
with pd.option_context('display.max_rows',10):
    display(df)
```

## Descrição textual dos dados

`sepal_length` representa o comprimento da pétala

`sepal_width` representa a largura da pétala

`petal_length` representa o comprimento da sépala

`petal_width` representa a largura da sépala

`species` representa as espécie/raça da planta

```python
df["species"].unique()
```

### Relação entre uma variável categórica e uma contínua

- Vizualização `sepal_length` em relação às espécies
- Vizualização `sepal_width` em relação às espécies
- Vizualização `petal_length` em relação às espécies
- Vizualização `petal_width` em relação às espécies

```python
fig, ax = plt.subplots(1,4, figsize=(16,6))
sns.catplot(x="species", y="sepal_length", data=df, ax=ax[0]);
sns.catplot(x="species", y="sepal_width" , data=df, ax=ax[1]);
sns.catplot(x="species", y="petal_length", data=df, ax=ax[2]);
sns.catplot(x="species", y="petal_width" , data=df, ax=ax[3]);
```

```python
fig, ax = plt.subplots(1,4, figsize=(16,6))
sns.catplot(x="species", y="sepal_length",ax=ax[0], kind="box",data=df);
sns.catplot(x="species", y="sepal_width", ax=ax[1], kind="box", data=df);
sns.catplot(x="species", y="petal_length",ax=ax[2], kind="box", data=df);
sns.catplot(x="species", y="petal_width", ax=ax[3], kind="box", data=df);
```
