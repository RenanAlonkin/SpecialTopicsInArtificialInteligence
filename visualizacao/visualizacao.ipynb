```python
x1 = np.random.normal(size=100)
sns.distplot(x1)
```

```python
x2 = np.random.normal(size=1000)
x3 = np.random.normal(size=10000)
x4 = np.random.normal(size=100000)

fig, ax = plt.subplots(1,3, figsize=(16,4))
sns.distplot(x2, ax=ax[0])
sns.distplot(x3, ax=ax[1])
sns.distplot(x4, ax=ax[2])
```

```python
#Dist Poisson, usado para distribuições como tempo.
x2 = np.random.poisson(3,size=1000)
x3 = np.random.poisson(5,size=10000)
x4 = np.random.poisson(10,size=100000)

fig, ax = plt.subplots(1,3, figsize=(16,4))

#KDE Kernel Density (linha que representa a distribuição)
sns.distplot(x2, ax=ax[0], bins=15, kde=False).set(xlim=(0,20))
sns.distplot(x3, ax=ax[1], bins=15, kde=False).set(xlim=(0,20))
sns.distplot(x4, ax=ax[2], bins=15, kde=False).set(xlim=(0,20))
```
