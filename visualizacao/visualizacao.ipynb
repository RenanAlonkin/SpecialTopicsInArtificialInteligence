```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = np.arange(10)
plt.plot(data1)
```

```python
x= np.arange(10)
y= x**2
plt.plot(x,y)
```

```python
fig1 = plt.figure(figsize=(14,8))
ax1=fig1.add_subplot(2,2,1)
ax2=fig1.add_subplot(2,2,2)
ax3=fig1.add_subplot(2,2,3)
```

```python
ax1.plot(data1)
ax2.plot(x,y)
ax3.scatter(x= np.arange(10), y= np.arange(10)+ np.random.randn(10))
fig1
```

```python
fig, axes = plt.subplots(2,3, figsize=(14,8))
for i in range(axes.shape[0]):
    for j in range(axes.shape[1]):
        axes[i,j].hist(np.random.randn(1000), bins=50, color='k', alpha=0.5)
```

```python
## Conjunto de dados tipos 
df = sns.load_dataset('tips')
df.head()
```

```python
# Mostrando os 5 elementos em cima e embaixo.
```

```python
with pd.option_context('display.max_rows', 7):
    display(df)
```

```python
sns.relplot(x='total_bill',y='tip', data=df)
```

```python
sns.relplot(x='total_bill',y='tip', hue='sex', data=df)
```

```python
sns.relplot(x='total_bill',y='tip', style='smoker',hue='sex', data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex', size='size', data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex', size='size', sizes=(20,200), data=df)
```

```python
## categorica e contínua
sns.catplot(x='day', y='total_bill', data=df)
```

```python
## categorica e contínua. Jitter diz a respeito a aproximação 
sns.catplot(x='day', y='total_bill',jitter=False, data=df)
```

```python
## Gráfico de barras
sns.catplot(x='day',y='total_bill',kind='bar', data=df)
```

```python
## Gráfico de barras
sns.catplot(x='smo',y='total_bill',kind='bar', data=df)
```
